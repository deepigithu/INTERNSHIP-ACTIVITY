BOWLER ANALYSIS

```python
df=pd.read_csv(r"C:\Users\Deepika\Downloads\Bowler_data.csv")
```

DATA AUDIT

```python
df.drop(df[df.Overs == '-'].index, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
#Replacing all the '-' with 0s to make the dtypes of a column uniform
df['Mdns'] = df['Mdns'].str.replace('-', '0')
df['Runs'] = df['Runs'].str.replace('-', '0')
df['Wkts'] = df['Wkts'].str.replace('-', '0')
df['Econ'] = df['Econ'].str.replace('-', '0')
df['Ave'] = df['Ave'].str.replace('-', '0')
df['SR'] = df['SR'].str.replace('-', '0')
```

```python
#Converting the dtypes of certain columns into float to perform operations
df['Overs'] = df['Overs'].astype(float)
df['Mdns'] = df['Mdns'].astype(float)
df['Runs'] = df['Runs'].astype(float)
df['Wkts'] = df['Wkts'].astype(float)
df['Econ'] = df['Econ'].astype(float)
df['Ave'] = df['Ave'].astype(float)
df['SR'] = df['SR'].astype(float)
df.dtypes
```

```python
#Dropping all the values with 'Did not Bat (DNB)'
bd.drop(bd[bd.Bat1 == 'DNB'].index, inplace=True)
bd.head()
```

```python
#To Convert required columns into float we have to replace and remove some string elements
bd['Bat1'] = bd['Bat1'].str.replace('*', '.')
bd['SR'] = bd['SR'].str.replace('-', '0')
bd['Runs'] = bd['Runs'].str.replace('-', '0')
bd['4s'] = bd['4s'].str.replace('-', '0')
bd['6s'] = bd['6s'].str.replace('-', '0')
bd.head()
```

```python
#Converting all required columns into float
bd['SR']=bd['SR'].astype(float)
bd['Runs']=bd['Runs'].astype(float)
bd['4s']=bd['4s'].astype(float)
bd['6s']=bd['6s'].astype(float)
```
